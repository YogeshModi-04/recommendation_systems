```python
import pandas as pd
import numpy as np
```

```python
data  = pd.read_csv("data/synthetic_reco_dataset_1000.csv")
df = data.copy()
df.head()
```

```python
df.dtypes
```

```python
df.drop(columns=['Brand_category'], inplace=True)
df.head()
```

```python
df.shape
```

```python
## check missing values
df.isnull().sum()
```

```python
## removing INR from cost column and converting to float
df["cost"] = (
    df["cost"]
    .str.replace(r"[^\d\.]", "", regex=True)   # remove “INR”, commas, etc.
    .astype(float)
)
```

```python
df.head()
```

```python
# Standardize text fields
for col in ["Gender", "Location", "Brand", "product_category", "category", "Gender_category","Product"]:
    df[col] = df[col].str.strip().str.title()
```

```python
df.dtypes
```

```python
# printing the unique values in each column value count
print("Unique values in each column:\n")
for col in df.columns:
    print(f"{col}: {df[col].nunique()}")
```

```python
# applying categorical encoding on Gender, Location, Brand, product_category, category, Gender_category
df['Gender'] = df['Gender'].astype("category").cat.codes
df['Location'] = df['Location'].astype("category").cat.codes
df['Brand'] = df['Brand'].astype("category").cat.codes
df['product_category'] = df['product_category'].astype("category").cat.codes
df['category'] = df['category'].astype("category").cat.codes
df['Gender_category'] = df['Gender_category'].astype("category").cat.codes  
```

```python
df.head()
```

```python
# generalising product id because we have only 43 products only
# will map product id to a range of 0-42
# we will loop through product and assign a new value to product id column
product_mapping = {product: idx for idx, product in enumerate(df['Product'].unique())}
print("Product mapping:", product_mapping)
```

```python
# will re assign the product id column with the new values
for product, idx in product_mapping.items(): # will replace product_id with idx
    # will get corresponding value of product_id and change it to idx
    df["product_id"] = df["Product"].replace(product, idx)
    
df['product_id'] = df['product_id'].astype("category").cat.codes 
df.head()
```

```python
# printing the unique values in each column value count
print("Unique values in each column:\n")
for col in df.columns:
    print(f"{col}: {df[col].nunique()}")
```

```python
df.drop(columns=['Product'], inplace=True)  # drop the original Product column
df.head()
```

```python
# printing the unique values in each column value count
print("Unique values in each column:\n")
for col in df.columns:
    print(f"{col}: {df[col].nunique()}")
```

```python
from sklearn.preprocessing import StandardScaler

# 0.1 Select and scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[['Age','Gender','Location']])
df[['Age_s','Gender_s','Location_s']] = X_scaled
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans

# 1. Build a one‐row‐per‐user feature table (they’ll all have identical Age/Gender/Location across their purchases)
user_feats = (
    df
    .groupby('user_id')[['Age_s','Gender_s','Location_s']]
    .first()          # or .mean() – since your demos are constant per user it doesn’t matter
    .reset_index()
)

# 2. Fit K-Means on those scaled demos
n_clusters = 10
kmeans    = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(user_feats[['Age_s','Gender_s','Location_s']])

# 3. Attach the cluster labels back
user_feats['cluster'] = kmeans.labels_
```

```python
# user_feats = DataFrame with user_id, Age_s, Gender_s, Location_s, cluster
df = df.merge(
    user_feats[['user_id','cluster']],
    on='user_id',
    how='left'
)
```

```python
# Here we’ll just use df and .value_counts()
cluster_top10 = (
    df
      .groupby('cluster')['product_id']
      .agg(lambda prods: prods.value_counts().head(10).index.tolist())
      .to_dict()
)
```

```python
def recommend_cluster(age, gender, location, scaler, kmeans, cluster_map, top_k=10):
    """
    Scale the inputs, predict the cluster, and return its top_k products.
    """
    # 3.1 Scale the triplet
    x = scaler.transform([[age, gender, location]])
    # 3.2 Predict cluster
    cl = kmeans.predict(x)[0]
    # 3.3 Return the top-k for that cluster (may be fewer if not enough products)
    return cluster_map.get(cl, [])[:top_k]
```

```python
# ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# Suppose a new user: age=32, gender=0, location=15
new_age, new_gender, new_loc = 32, 0, 15

recs = recommend_cluster(
    age=new_age,
    gender=new_gender,
    location=new_loc,
    scaler=scaler,
    kmeans=kmeans,
    cluster_map=cluster_top10,
    top_k=10
)
print("Cluster-based recommendations:", recs)
```
